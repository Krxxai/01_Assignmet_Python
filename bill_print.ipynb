```python
def take_order(tbl_no, item, qnt):
    tbl_dict.update({tbl_no:{item:qnt}})
    return tbl_dict
```

```python
def add_order (ord_num,item,qnt):
    key1= list(ord_num.keys())[0]
    ord_num[key1].update({item:qnt})
    return ord_num
```

```python

tbl_dict={}
tbl_no=input("")
while True:
    next= input("""" Press
                        1 to add new table order
                        2 to add old table next order
                        0 to exit """)

    if next=='0':
        break
    elif next == '1':
        item =input("")
        qnt = input()
        ord_out= take_order(tbl_no, item ,qnt)
        print(ord_out)
    elif next == '2':
        item =input("")
        qnt = input()
        tot_ord= add_order(ord_out, item ,qnt)
        print(tot_ord)
    else:
        print("wrong output")
```
